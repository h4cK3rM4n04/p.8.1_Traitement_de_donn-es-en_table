# Placer une ville sur une carte.
## 1 . Objectif.
l'objectif de cette activité est de placer sur une carte une ville dont le code postal est fourni en paramètre.
## 2. La bibliothèque `folium`.
Folium est une bibliothèque Python, permettant de générer des cartes, et de les afficher. En se basant sur le coordonnées GPS, il est possible de placer des marqueurs sur ces cartes.
Un petit exemple :

```python
import folium
import csv

carte = folium.Map(location=[43,2], zoom_start=7)

folium.Marker([43.4947364798,1.37416616378],radius=10, popup ='ROQUETTES').add_to(carte)

carte
```

## 3.Travail à réaliser
**Dans la cellule de code ci-dessous, vous devez écrire un script qui affiche la localisation d'une ville sur une carte à partir de son code postal passé en paramètre**

```python
import csv
param = int(input())
file_csv = "commune_valide.csv"
with open(file_csv) as hey:
    cpt = 0
    for x in hey:
        if cpt >= 1:
            i = x.split(";")
            ent = i[2] 
            print(ent)
        cpt += 1
```

```python
import folium
import csv
param = input("Entrez le code postal de la ville : ")
file_csv = "commune_valide.csv"

with open(file_csv) as hey:
    reader = csv.DictReader(hey, delimiter=';')
    for row in reader:
        if row['Code_postal'] == param:
            coord = row['coordonnees_gps'].split(',')
            latitude = float(coord[0])
            longitude = float(coord[1])
            nom_ville = row['Nom_commune']
            break
    else:
        print("Aucune ville trouvée avec ce code postal.")

#Création de la carte avec les coordonnées de la ville
map = folium.Map(location=[latitude, longitude], zoom_start=13)

# Ajout d'un marqueur à l'emplacement de la ville sur la carte
folium.Marker(location=[latitude, longitude], popup=nom_ville).add_to(map)
map
```
